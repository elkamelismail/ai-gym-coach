```python
import os
from langchain_community.document_loaders import PyPDFLoader
```

```python
raw_docs_paths = os.path.abspath("../documents/raw")
cats = os.listdir(raw_docs_paths)

for cat in cats:
    print("-" * 50)
    print(f"category: {cat}")
    cat_path = os.path.join(raw_docs_paths, cat)
    docs = os.listdir(cat_path)
    print(f"documents count: {len(docs)}")
```

```python
processed_docs = os.path.abspath("../documents/processed")

for cat in cats:
    cat_path = os.path.join(raw_docs_paths, cat)
    processed_cat_path = os.path.join(processed_docs, cat)
    if not os.path.exists(processed_cat_path):
        os.makedirs(processed_cat_path)
    docs = os.listdir(cat_path)
    for doc in docs:
        doc_path = os.path.join(cat_path, doc)

        if not doc.endswith(".pdf"):
            continue
        
        loader = PyPDFLoader(doc_path)
        pages = loader.load()
        processed_doc_path = os.path.join(processed_cat_path, f"{doc[:-4]}.md")
        with open(processed_doc_path, "w") as f:
            for page in pages:
                f.write(page.page_content)
                f.write("\n\n")
        print(f"Document {doc} in category {cat} processed successfully")
                
                
```
